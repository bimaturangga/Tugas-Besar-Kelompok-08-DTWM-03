```python
import pandas as pd

print("Memulai proses ekstraksi data...\n")

print("=== SOURCE 1: GTFS ===")
stops = pd.read_csv('stops.txt')
stop_times = pd.read_csv('stop_times.txt')

print(">>> stops.txt")
print(f"Shape: {stops.shape}")
print("-" * 30)
stops.info()
display(stops.head())

print("\n>>> stop_times.txt")
print(f"Shape: {stop_times.shape}")
print("-" * 30)
stop_times.info()
display(stop_times.head())

print("\n=== SOURCE 2: MANUAL DATA ===")
fares_ops = pd.read_csv('Data Operasional Transjakarta - Data Operasional Transjakarta.csv')

print(">>> fares_ops (Data Operasional Transjakarta)")
print(f"Shape: {fares_ops.shape}")
print("-" * 30)
fares_ops.info()
display(fares_ops.head())

print("\n=== SOURCE 3: API CALL (GMAPS) ===")
ratings = pd.read_csv('transjakarta_synthetic.csv')

print(">>> ratings (Google Maps Halt Ratings)")
print(f"Shape: {ratings.shape}")
print("-" * 30)
ratings.info()
display(ratings.head())

print("\nProses ekstraksi selesai.")
```

```python
import numpy as np

print("1. Menangani Missing Values...")
stops.dropna(how='all', axis=1, inplace=True)
if 'zone_id' in stops.columns:
    stops['zone_id'] = stops['zone_id'].fillna('UNKNOWN')

ratings['rating'] = ratings['rating'].fillna(ratings['rating'].median())
ratings['user_ratings_total'] = ratings['user_ratings_total'].fillna(0)


print("2. Standarisasi Nama Halte...")
stops['stop_name_clean'] = stops['stop_name'].astype(str).str.upper().str.strip()
ratings['stop_name_clean'] = ratings['stop_name'].astype(str).str.upper().str.strip()

stops_enriched = pd.merge(stops, ratings[['stop_name_clean', 'rating']],
                          on='stop_name_clean', how='left')

print("3. Standarisasi Waktu Operasional...")
def fix_gtfs_time(time_str):
    """Mengubah format GTFS (misal 25:15:00) menjadi format standar 24 jam (01:15:00)"""
    if pd.isna(time_str):
        return time_str
    h, m, s = map(int, str(time_str).split(':'))
    h_fixed = h % 24
    return f"{h_fixed:02d}:{m:02d}:{s:02d}"

def time_to_minutes(time_str):
    if pd.isna(time_str): return np.nan
    h, m, s = map(int, str(time_str).split(':'))
    return (h * 60) + m + (s / 60.0)

stop_times['arrival_time_fixed'] = stop_times['arrival_time'].apply(fix_gtfs_time)
stop_times['departure_time_fixed'] = stop_times['departure_time'].apply(fix_gtfs_time)
stop_times['arrival_minutes_raw'] = stop_times['arrival_time'].apply(time_to_minutes)


print("4. Membentuk Tabel Dimensi...")

# --- DIM_STOP ---
dim_stop = stops_enriched[['stop_id', 'stop_name', 'stop_lat', 'stop_lon', 'zone_id', 'rating']].copy()
dim_stop.drop_duplicates(subset=['stop_id'], inplace=True)
dim_stop.insert(0, 'stop_sk', range(1, len(dim_stop) + 1)) # Create Surrogate Key

# --- DIM_ROUTE ---
stop_times['route_id'] = stop_times['trip_id'].str.split('-').str[0]
dim_route = pd.DataFrame({'route_id': stop_times['route_id'].unique()})
dim_route.insert(0, 'route_sk', range(1, len(dim_route) + 1))

# --- DIM_OPERATOR ---
dim_operator = pd.DataFrame({
    'operator_sk': [1],
    'agency_id': ['Tije'],
    'agency_name': ['Transjakarta'],
    'transport_mode': ['BRT & Non-BRT']
})
stop_times['operator_sk'] = 1 # Karena seluruh data diasumsikan milik Transjakarta

# --- DIM_TIME ---
unique_times = pd.concat([stop_times['arrival_time_fixed'], stop_times['departure_time_fixed']]).dropna().unique()
dim_time = pd.DataFrame({'time_string': unique_times})
dim_time.insert(0, 'time_sk', range(1, len(dim_time) + 1))

# Derived Columns untuk Waktu (hour_bin & day_type)
dim_time['hour_of_day'] = dim_time['time_string'].str.split(':').str[0].astype(int)

def get_hour_bin(h):
    if 5 <= h < 9: return 'Morning Peak'
    elif 9 <= h < 16: return 'Off-Peak'
    elif 16 <= h < 20: return 'Evening Peak'
    elif 20 <= h < 24: return 'Night'
    else: return 'Midnight / AMARI'

dim_time['hour_bin'] = dim_time['hour_of_day'].apply(get_hour_bin)

dim_time['day_type'] = 'Weekday'


print("5. Membentuk Fact Table & Menghitung Headway...\n")

fact_df = stop_times.copy()

# A. Mapping Foreign Keys (Surrogate Keys) ke tabel fakta
fact_df = fact_df.merge(dim_stop[['stop_id', 'stop_sk']], on='stop_id', how='left')
fact_df = fact_df.merge(dim_route[['route_id', 'route_sk']], on='route_id', how='left')

# Role-playing dimension mapping untuk Waktu Tiba dan Waktu Berangkat
fact_df = fact_df.merge(dim_time[['time_string', 'time_sk']], left_on='arrival_time_fixed', right_on='time_string', how='left')
fact_df.rename(columns={'time_sk': 'arrival_time_sk'}, inplace=True)

fact_df = fact_df.merge(dim_time[['time_string', 'time_sk']], left_on='departure_time_fixed', right_on='time_string', how='left')
fact_df.rename(columns={'time_sk': 'departure_time_sk'}, inplace=True)

# B. Hitung Derived Column: Headway (Selisih waktu kedatangan antar armada di rute dan halte yang sama)
# Sort berdasarkan route, stop_id, dan waktu kedatangan ASLI (agar jadwal setelah jam 24:00 tidak berantakan)
fact_df = fact_df.sort_values(by=['route_id', 'stop_id', 'arrival_minutes_raw'])

# Menghitung selisih menit dengan kedatangan sebelumnya (.diff())
fact_df['headway_minutes'] = fact_df.groupby(['route_id', 'stop_id'])['arrival_minutes_raw'].diff()
fact_df['headway_minutes'] = fact_df['headway_minutes'].round(1)

# C. Finalisasi Kolom Fact Table
fact_columns = [
    'route_sk', 'operator_sk', 'stop_sk', 'arrival_time_sk', 'departure_time_sk',
    'trip_id', 'stop_sequence', 'shape_dist_traveled', 'headway_minutes'
]
fact_trip_schedules = fact_df[fact_columns].copy()
fact_trip_schedules.insert(0, 'trip_schedule_sk', range(1, len(fact_trip_schedules) + 1))

print(">>> PREVIEW FACT TABLE (fact_trip_schedules):")
display(fact_trip_schedules.head())

print("\n>>> PREVIEW DIM TIME (dengan hour_bin & day_type):")
display(dim_time.head())
```

```python
from google.cloud import bigquery
from google.oauth2 import service_account

print("==========================================")
print("B3 — LOAD: Memasukkan Data ke BigQuery")
print("==========================================\n")

# Setup Credentials & Client
# Pastikan file 'credentials.json' ada di direktori yang sama
KEY_PATH = 'credentials.json'
PROJECT_ID = 'tubes1-bigquery' # UBAH INI
DATASET_ID = 'transjakarta_dwh' # UBAH INI (Pastikan dataset ini sudah dibuat di BQ)

credentials = service_account.Credentials.from_service_account_file(KEY_PATH)
client = bigquery.Client(credentials=credentials, project=PROJECT_ID)

# Job Config untuk replace tabel jika sudah ada (bisa diganti 'WRITE_APPEND' jika ingin menambah data)
job_config = bigquery.LoadJobConfig(write_disposition="WRITE_TRUNCATE")

# Dictionary tabel yang akan di-load
tables_to_load = {
    'dim_stop': dim_stop,
    'dim_route': dim_route,
    'dim_operator': dim_operator,
    'dim_time': dim_time,
    'fact_trip_schedules': fact_trip_schedules
}

# Proses Looping Load ke BQ
for table_name, df in tables_to_load.items():
    table_id = f"{PROJECT_ID}.{DATASET_ID}.{table_name}"
    print(f"Loading data into {table_id}...")

    # Eksekusi Load
    job = client.load_table_from_dataframe(df, table_id, job_config=job_config)
    job.result() # Tunggu hingga proses selesai

    print(f"✅ Berhasil! {job.output_rows} baris dimuat ke tabel {table_name}.\n")

print("Proses Load ke BigQuery Selesai!")
```

```python
#BUSINESS QUESTION 5
WITH StopStats AS (
    -- Langkah 1: Hitung jumlah rute untuk masing-masing halte terlebih dahulu
    SELECT
        ds.stop_id,
        ds.rating,
        COUNT(DISTINCT fts.route_sk) AS total_rute
    FROM
        `tubes1-bigquery.transjakarta_dwh.fact_trip_schedules` fts
    JOIN
        `tubes1-bigquery.transjakarta_dwh.dim_stop` ds
        ON fts.stop_sk = ds.stop_sk
    WHERE
        ds.rating IS NOT NULL
    GROUP BY
        ds.stop_id, ds.rating
)

-- Langkah 2: Kelompokkan berdasarkan jumlah rute yang sudah dihitung
SELECT
    CASE
        WHEN total_rute >= 10 THEN '>= 10 Rute'
        WHEN total_rute BETWEEN 7 AND 9 THEN '7 - 9 Rute'
        WHEN total_rute BETWEEN 4 AND 6 THEN '4 - 6 Rute'
        ELSE '1 - 3 Rute'
    END AS kategori_jumlah_rute,
    ROUND(AVG(rating), 2) AS rata_rata_rating,
    COUNT(DISTINCT stop_id) AS jumlah_halte_dalam_kategori
FROM
    StopStats
GROUP BY
    kategori_jumlah_rute
ORDER BY
    rata_rata_rating DESC;
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

stops = pd.read_csv('stops.txt')
stop_times = pd.read_csv('stop_times.txt')
routes = pd.read_csv('routes.txt')
trips = pd.read_csv('trips.txt')
calendar = pd.read_csv('calendar.txt')
ratings = pd.read_csv('transjakarta_synthetic.csv')

print("Data berhasil dimuat! Lanjut ke cell berikutnya.")
```

```python
#Heatmap kepadatan jadwal

stop_times['hour'] = stop_times['arrival_time'].str.split(':').str[0].astype(float) % 24

trip_cal = trips.merge(calendar, on='service_id', how='left')
st_trip = stop_times.merge(trip_cal[['trip_id', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']], on='trip_id', how='inner')

days = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
day_names = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']

melted = st_trip.melt(id_vars=['hour'], value_vars=days, var_name='day', value_name='active')
active_schedules = melted[melted['active'] == 1]
heatmap_data = active_schedules.groupby(['day', 'hour']).size().reset_index(name='count')
heatmap_pivot = heatmap_data.pivot(index='day', columns='hour', values='count').fillna(0)

heatmap_pivot.index = pd.CategoricalIndex(heatmap_pivot.index, categories=days, ordered=True)
heatmap_pivot = heatmap_pivot.sort_index()
heatmap_pivot.index = day_names

plt.clf()
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(heatmap_pivot, cmap="YlOrRd", annot=False, ax=ax)

ax.set_title("Kepadatan Jadwal Transjakarta: Hari vs Jam Operasional", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Jam Operasional (0-23)", fontsize=12)
ax.set_ylabel("Hari dalam Seminggu", fontsize=12)

plt.tight_layout()
plt.savefig("heatmap.png")
plt.show()
```

```python
# Scatter Plot Rute vs Rating

stops['stop_name_clean'] = stops['stop_name'].astype(str).str.upper().str.strip()
ratings['stop_name_clean'] = ratings['stop_name'].astype(str).str.upper().str.strip()

st_tr = stop_times.merge(trips[['trip_id', 'route_id']], on='trip_id', how='left')
routes_per_stop = st_tr.groupby('stop_id')['route_id'].nunique().reset_index(name='num_routes')

stops_enriched = stops.merge(ratings[['stop_name_clean', 'rating']], on='stop_name_clean', how='inner')
stops_enriched = stops_enriched.merge(routes_per_stop, on='stop_id', how='left')
stops_enriched['rating'] = stops_enriched['rating'].fillna(stops_enriched['rating'].median())
stops_enriched['num_routes'] = stops_enriched['num_routes'].fillna(1)

plt.clf()
fig, ax = plt.subplots(figsize=(10, 6))

sns.scatterplot(data=stops_enriched, x='num_routes', y='rating', alpha=0.6, s=100, color='#2F489C', ax=ax)
sns.regplot(data=stops_enriched, x='num_routes', y='rating', scatter=False, color='#D62126', ax=ax)

ax.set_title("Korelasi Jumlah Rute yang Melayani Halte terhadap Rating Maps", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Jumlah Rute yang Melayani Halte", fontsize=12)
ax.set_ylabel("Rating Google Maps (1-5)", fontsize=12)

plt.tight_layout()
plt.savefig("scatter.png")
plt.show()
```

```python
# Bar Chart

def time_to_minutes(time_str):
    if pd.isna(time_str): return np.nan
    parts = str(time_str).split(':')
    if len(parts) != 3: return np.nan
    h, m, s = map(int, parts)
    return (h * 60) + m + (s / 60.0)

st_tr['arrival_mins'] = st_tr['arrival_time'].apply(time_to_minutes)
st_tr = st_tr.sort_values(by=['route_id', 'stop_id', 'arrival_mins'])
st_tr['headway'] = st_tr.groupby(['route_id', 'stop_id'])['arrival_mins'].diff()

valid_hw = st_tr[(st_tr['headway'] > 0) & (st_tr['headway'] < 120)]
avg_hw = valid_hw.groupby('route_id')['headway'].mean().reset_index()
avg_hw = avg_hw.merge(routes[['route_id', 'route_short_name']], on='route_id', how='left')
avg_hw = avg_hw.sort_values('headway').head(15)

plt.clf()
fig, ax = plt.subplots(figsize=(12, 6))

sns.barplot(data=avg_hw, x='route_short_name', y='headway', palette="viridis", ax=ax)

ax.set_title("15 Rute dengan Rata-Rata Waktu Tunggu (Headway) Tercepat", fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel("Kode Rute Transjakarta", fontsize=12)
ax.set_ylabel("Estimasi Waktu Tunggu (Menit)", fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("bar.png")
plt.show()
```
